Backtrack

```python
def BT():
    def bt(v, gran_cuenta):
        L = len(v)
        check = {3:(0,1,2), 6:(3,4,5), 7:(0,3,6), 8:(1,4,7), 9:(2,5,8)}
        if L in check.keys():
            if sum(v[i] for i in check[L]) != 18:
                return gran_cuenta
        if L==3:
            print(v)
        if L==9:
            gran_cuenta += 1
            # print(v)
            return gran_cuenta
        candidatos = list(range(1,13))
        for e in candidatos:
            v.append(e)
            gran_cuenta = bt(v, gran_cuenta)
            v.pop()
        return gran_cuenta
    gran_cuenta = bt([], 0)
    return gran_cuenta
```
