```python
def count_letters(input_txt):
    counts = {}
    input_txt = input_txt.translate(input_txt.maketrans('', '', '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~•')).lower() # Removes punctuation marks
    words = input_txt.split()
    for word in words:
        for letter in word:
            counts[letter] = counts.get(letter, 0) + 1
    # To sort dictionary by descending values
    sorting_lst = []
    for letter, count in counts.items():
        sorting_lst.append((count, letter))
    sorting_lst.sort(reverse=True)
    # Prints dict
    for count, letter in sorting_lst:
        print(f"{letter:4} {count}")
```

```python
count_letters('últimamente hemos tenido varios días muy lluviosos')
```

```python
def count_words(input_lst):
    counts = {}
    input_txt = ' '.join(input_lst)
    input_txt = input_txt.translate(input_txt.maketrans('', '', '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~•')).lower() # Removes punctuation marks
    words = input_txt.split()
    for word in words:
        counts[word] = counts.get(word, 0) + 1
    # To sort dictionary by ascending keys
    counts = sorted(counts.items())
    # Prints dict
    for word, count in counts:
        print(f"{word:18} {count}")
```

```python
txt = ['Hola , cómo estás!', 
       'gana dinero , gana desde casa.',
       'Hola , puedes llamar ahora?',
       'Hola , te puedo llamar mañana?']
```

```python
count_words(txt)
```
